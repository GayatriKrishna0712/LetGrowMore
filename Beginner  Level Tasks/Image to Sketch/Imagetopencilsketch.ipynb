1. Convert RGB color image to grayscale
2. Invert the grayscale image to a negative
3. Apply Gaussian blur to the negative
4. Blend the gray scale image from 1 with the blurred negative using color dodge

```python
from google.colab.patches import cv2_imshow
import cv2
print(cv2.__version__)
```

```python
#load the image
image = cv2.imread("th.jpeg")
cv2_imshow(image)
```

```python
#converting to gray scale
img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
#Invert the grayscale image to a negative
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

```python
#Apply Gaussian blur to the negative
img_blur = cv2.GaussianBlur(img_invert,(21,21),sigmaX = 0,sigmaY=0)
cv2_imshow(img_blur)
```

```python
#Blend the gray scale image from 1 with the blurred negative using color dodge
def dodgeV2(x,y):
    return cv2.divide(x,255-y,scale = 256)
```

```python
final_img = dodgeV2(img_gray,img_blur)
cv2_imshow(final_img)
```
